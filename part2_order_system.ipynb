```python

# ==============================================================
# TASK 2 — Cart Operations
# ==============================================================

print("\n" + "*" * 50)
print("TASK 2 — Cart Operations")
print("*" * 50)

cart = []


# ---- HELPER FUNCTIONS ----

def print_cart(cart):
    """Print the current cart contents in a readable format."""
    if len(cart) == 0:
        print("  [Cart is empty]")
    else:
        for entry in cart:
            # Each entry is like: {"item": "Paneer Tikka", "quantity": 2, "price": 180.0}
            line_total = entry["quantity"] * entry["price"]
            print(f"  {entry['item']:<20} x{entry['quantity']}   ₹{line_total:.2f}")


def add_to_cart(cart, item_name, quantity):
    """
    Add an item to the cart.
    - Checks if item exists in menu
    - Checks if item is available
    - If already in cart, increases quantity instead of duplicating
    """
    # Check 1: Does this item exist in the menu?
    if item_name not in menu:
        print(f"  ❌ '{item_name}' does not exist in the menu.")
        return  # 'return' exits the function early

    # Check 2: Is the item currently available?
    if not menu[item_name]["available"]:
        print(f"  ❌ '{item_name}' is currently unavailable.")
        return

    # Check 3: Is this item already in the cart?
    # We search through cart entries to find a matching item name
    for entry in cart:
        if entry["item"] == item_name:
            # Found it! Just increase the quantity
            entry["quantity"] += quantity
            print(f"  ✓ Updated '{item_name}' quantity to {entry['quantity']}.")
            return

    # If we get here, item is not yet in the cart — add a new entry
    cart.append({
        "item":     item_name,
        "quantity": quantity,
        "price":    menu[item_name]["price"]
    })
    print(f"  ✓ Added '{item_name}' x{quantity} to cart.")


def remove_from_cart(cart, item_name):
    """Remove an item completely from the cart."""
    for i, entry in enumerate(cart):  # enumerate gives us index AND value
        if entry["item"] == item_name:
            cart.pop(i)  # .pop(i) removes the item at position i
            print(f"  ✓ Removed '{item_name}' from cart.")
            return
    print(f"  ⚠ '{item_name}' is not in the cart.")


def update_quantity(cart, item_name, new_quantity):
    """Change the quantity of an item already in the cart."""
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] = new_quantity
            print(f"  ✓ Updated '{item_name}' quantity to {new_quantity}.")
            return
    print(f"  ⚠ '{item_name}' is not in the cart.")


# ---- SIMULATE THE SEQUENCE ----

print("\n--- Step 1: Add 'Paneer Tikka' x2")
add_to_cart(cart, "Paneer Tikka", 2)
print_cart(cart)

print("\n--- Step 2: Add 'Gulab Jamun' x1")
add_to_cart(cart, "Gulab Jamun", 1)
print_cart(cart)

print("\n--- Step 3: Add 'Paneer Tikka' x1")
add_to_cart(cart, "Paneer Tikka", 1)
print_cart(cart)

print("\n--- Step 4: Add 'Mystery Burger'")
add_to_cart(cart, "Mystery Burger", 1)
print_cart(cart)

print("\n--- Step 5: Add 'Chicken Wings'")
add_to_cart(cart, "Chicken Wings", 1)
print_cart(cart)

print("\n--- Step 6: Remove 'Gulab Jamun'")
remove_from_cart(cart, "Gulab Jamun")
print_cart(cart)

# ---- FINAL ORDER SUMMARY ----
print("\n========== Order Summary ==========")
subtotal = 0
for entry in cart:
    line_total = entry["quantity"] * entry["price"]
    subtotal  += line_total
    print(f"{entry['item']:<20} x{entry['quantity']}    ₹{line_total:.2f}")

gst   = round(subtotal * 0.05, 2)   # 5% GST
total = round(subtotal + gst, 2)

print("------------------------------------")
print(f"{'Subtotal:':<28} ₹{subtotal:.2f}")
print(f"{'GST (5%):':<28} ₹{gst:.2f}")
print(f"{'Total Payable:':<28} ₹{total:.2f}")
print("====================================")
```

```python

# ==============================================================
# TASK 4 — Daily Sales Log Analysis
# ==============================================================

print("\n" + "*" * 50)
print("TASK 4 — Daily Sales Log Analysis")
print("*" * 50)

# ---- PART A: Revenue per day ----
def print_revenue_table(log):
    """Compute and print revenue per day, and return the best day."""
    print(f"\n{'Date':<14} | {'Total Revenue':>14}")
    print("-" * 32)
    best_day    = None
    best_revenue = 0

    for date, orders in log.items():
        # Sum up all 'total' values across orders on this date
        day_revenue = sum(order["total"] for order in orders)
        print(f"  {date:<12} | ₹{day_revenue:>10.2f}")

        if day_revenue > best_revenue:
            best_revenue = day_revenue
            best_day     = date

    print(f"\nBest-selling day: {best_day} — ₹{best_revenue:.2f}")
    return best_day

print_revenue_table(sales_log)

# ---- PART B: Most ordered item ----
item_order_count = {}  # start with empty dict

for date, orders in sales_log.items():
    for order in orders:
        for item in order["items"]:

            if item not in item_order_count:
                item_order_count[item] = 0
            item_order_count[item] += 1

most_ordered_item  = max(item_order_count, key=lambda x: item_order_count[x])
most_ordered_count = item_order_count[most_ordered_item]
print(f"\nMost ordered item: '{most_ordered_item}' (appeared in {most_ordered_count} orders)")

# ---- PART C: Add new day and reprint ----
print("\n--- Adding new day: 2025-01-05 ---")
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

print("\nUpdated revenue table:")
print_revenue_table(sales_log)

# ---- PART D: Numbered list of ALL orders using enumerate ----
print("\n--- All Orders (across all dates) ---")
order_number = 1

for date, orders in sales_log.items():
    for order in orders:

        items_str = ", ".join(order["items"])
        print(f"  {order_number}. [{date}] Order #{order['order_id']:<3} — ₹{order['total']:.2f} — Items: {items_str}")
        order_number += 1
```
